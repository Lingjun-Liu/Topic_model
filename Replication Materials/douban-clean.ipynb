## step1 è¯»å–æ•°æ®

```python
import pandas as pd

# è¯»å– CSV æ–‡ä»¶
file_path = '/kaggle/input/douban-csv/final_data2.csv'
df = pd.read_csv(file_path)

# å¿«é€ŸæŸ¥çœ‹æ•°æ®è¡¨çš„æ¦‚å†µï¼Œç±»ä¼¼äº R ä¸­çš„ glimpse å‡½æ•°
df.info()

# æŸ¥çœ‹å‰å‡ è¡Œæ•°æ®ä»¥äº†è§£è¡¨çš„å†…å®¹
df.head()
```

## step2 åˆå¹¶è¯¥æœ‰çš„æ•°æ®æ 

```python

# ä» df ä¸­é€‰å‡ºï¼šæ ‡é¢˜ã€è¯„è®ºå†…å®¹ã€è¯„è®ºæ—¶é—´ã€è¯„è®ºipï¼Œç»„æˆ df1
df1 = df[['æ ‡é¢˜', 'è¯„è®ºå†…å®¹', 'è¯„è®ºæ—¶é—´', 'è¯„è®ºip']]
df1 = df1.rename(columns={'è¯„è®ºå†…å®¹': 'text', 'è¯„è®ºæ—¶é—´': 'year', 'è¯„è®ºip': 'ip'})
df1['year'] = df1['year'].astype(str).str[:4]
df1
```

```python
# ä» df ä¸­é€‰å‡ºï¼šæ ‡é¢˜ã€å†…å®¹ã€å¸–å­å‘å¸ƒæ—¶é—´ã€å¸–å­ipåœ°å€ï¼Œç»„æˆ df2
df2 = df.loc[:, ['æ ‡é¢˜', 'å†…å®¹', 'å¸–å­å‘å¸ƒæ—¶é—´', 'å¸–å­ipåœ°å€']].copy()
df2['text'] = df2['æ ‡é¢˜'] + df2['å†…å®¹']
df2 = df2.drop(columns=['å†…å®¹'])
df2 = df2.rename(columns={'å¸–å­å‘å¸ƒæ—¶é—´': 'year', 'å¸–å­ipåœ°å€': 'ip'})
df2['year'] = df2['year'].astype(str).apply(lambda x: x[:4] if x[:2] == '20' else '2024')
df2 = df2.drop_duplicates(subset='æ ‡é¢˜', keep='first')
df2
```

```python
# åˆå¹¶ df1 å’Œ df2ï¼Œå–å…¬å…±åˆ—ï¼šæ ‡é¢˜ã€yearã€ipã€text
data = pd.merge(df1, df2, on=['æ ‡é¢˜', 'year', 'ip'], how='outer', suffixes=('_df1', '_df2'))
data['text'] = data['text_df1'].combine_first(data['text_df2'])
data = data[['æ ‡é¢˜', 'year', 'ip', 'text']]

data
```

```python
# æŸ¥çœ‹ year åˆ—ä¸­æœ‰å“ªäº›å€¼
year_values = data['year'].value_counts()
print("\nYear values in the 'year' column:\n")
print(year_values)
```

```python
# æŸ¥çœ‹ ipåˆ—ä¸­æœ‰å“ªäº›å€¼
ip_values = data['ip'].value_counts()
print("\nYear values in the 'ip' column:\n")
print(ip_values)
# å¯¼å‡º ip_values ä¸º Excel æ–‡ä»¶
ip_values.to_excel('/kaggle/working/ip_values.xlsx', index=True)
```

```python
# æ–°å»ºä¸€åˆ— diquï¼Œæ ¹æ® ip åˆ—ä¸­çš„å€¼è¿›è¡Œç¼–ç 
import numpy as np
# æ–°å»ºä¸€åˆ— diquï¼Œæ ¹æ® ip åˆ—ä¸­çš„å€¼è¿›è¡Œç¼–ç 
def map_region(ip):
    eastern = ["å¹¿ä¸œ", "åŒ—äº¬", "æ±Ÿè‹", "æµ™æ±Ÿ", "ä¸Šæµ·", "å±±ä¸œ", "æ²³åŒ—", "ç¦å»º", "å¤©æ´¥", "æµ·å—", "å¹¿è¥¿", "ä¸­å›½é¦™æ¸¯", "ä¸­å›½æ¾³é—¨", "ä¸­å›½å°æ¹¾"]
    central = ["å››å·", "è¾½å®", "æ¹–å—", "æ¹–åŒ—", "æ²³å—", "å®‰å¾½", "é‡åº†", "å‰æ—", "é™•è¥¿", "æ±Ÿè¥¿"]
    western = ["äº‘å—", "å±±è¥¿", "å†…è’™å¤", "è´µå·", "ç”˜è‚ƒ", "æ–°ç–†", "é’æµ·", "è¥¿è—"]
    overseas = ["ç¾å›½", "æ—¥æœ¬", "å¾·å›½", "åŠ æ‹¿å¤§", "æ¾³å¤§åˆ©äºš", "æ–°åŠ å¡", "è·å…°", "è‹±å›½", "æ„å¤§åˆ©", "éŸ©å›½", "æ³•å›½", "æ¯”åˆ©æ—¶", "æ–°è¥¿å…°", "æ³°å›½", "çˆ±å°”å…°", "ä¸¹éº¦", "è¶Šå—", "é©¬æ¥è¥¿äºš", "å°¼æ³Šå°”", "ä¿„ç½—æ–¯", "å§”å†…ç‘æ‹‰", "è¥¿ç­ç‰™", "å°åº¦å°¼è¥¿äºš", "é˜¿è”é…‹", "æŒªå¨", "ç‘å£«", "ç™½ä¿„ç½—æ–¯", "å…‹ç½—åœ°äºš", "ç‘å…¸", "å‡ å†…äºš"]
    
    if ip in eastern:
        return 1
    elif ip in central:
        return 2
    elif ip in western:
        return 3
    elif ip in overseas:
        return 4
    else:
        return np.nan

data.loc[:, 'diqu'] = data['ip'].apply(map_region)
```

## step4 æ¸…æ´—ç‰¹æ®Šå­—ç¬¦

```python
import re
# å®šä¹‰æ¸…æ´—å‡½æ•°
def clean_text(text):
    # ç¬¬ä¸€æ­¥ï¼šåˆ é™¤æ— ç”¨å­—ç¬¦ï¼ˆç©ºæ ¼ã€tabé”®ã€æ¢è¡Œç¬¦ã€å›è½¦ç¬¦ï¼‰
    text = text.replace(' ', '').replace('\t', '').replace('\n', '').replace('\r', '')
    
    # ç¬¬äºŒæ­¥ï¼šåˆ é™¤æ ‡ç‚¹ç¬¦å·ï¼ˆä¿ç•™å­—æ¯ã€æ•°å­—å’Œä¸‹åˆ’çº¿ï¼‰
    text = re.sub(r'[^\w]', '', text)
    
    # ç¬¬ä¸‰æ­¥ï¼šä»…æå–ä¸­æ–‡æ–‡æœ¬
    text = re.sub(r'[^\u4e00-\u9fa5]', '', text)
    
    return text

# å¯¹ 'text' åˆ—è¿›è¡Œæ¸…æ´—ï¼Œå¹¶å°†ç»“æœå­˜å‚¨åˆ°æ–°çš„ 'text1' åˆ—
data['text1'] = data['text'].apply(lambda x: clean_text(str(x)))

# æŸ¥çœ‹å¤„ç†åçš„æ•°æ®
data[['text', 'text1']]
```

```python
# åˆ é™¤ 'text1' åˆ—ä¸­ä¸ºç©ºçš„æ•°æ®è¡Œ
data = data[data['text1'] != '']

# æŸ¥çœ‹å¤„ç†åçš„æ•°æ®
data[['text', 'text1']]
```

## step5 æ¸…æ´—å†…å®¹ä¸å¯è§

```python
# åˆ é™¤ 'text1' åˆ—ä¸­å†…å®¹ä¸º "å†…å®¹ä¸å¯è§" çš„æ•°æ®è¡Œ
data = data[data['text1'] != 'å†…å®¹ä¸å¯è§']

# æŸ¥çœ‹å¤„ç†åçš„æ•°æ®
data[['text', 'text1']]
```

## step6 åˆ†è¯&ä¿ç•™ç‰¹å®šè¯æ€§çš„è¯

```python
import pandas as pd
import jieba.posseg as pseg

# å®šä¹‰åˆ†è¯å‡½æ•°ï¼Œåªä¿ç•™åè¯(n)ã€åŠ¨è¯(v)å’Œå½¢å®¹è¯(a)
def jieba_cut_filtered(text):
    # ä½¿ç”¨ jieba.posseg è¿›è¡Œåˆ†è¯å¹¶è·å–è¯æ€§
    words = pseg.cut(text)
    # åªä¿ç•™åè¯(n)ã€åŠ¨è¯(v)ã€å½¢å®¹è¯(a)çš„è¯,é‡è¯ï¼Œæ•°è¯
    filtered_words = [word for word, flag in words if flag in ('n', 'v', 'a','m','q')]
    return ' '.join(filtered_words)

# å¯¹ 'text1' åˆ—è¿›è¡Œåˆ†è¯ï¼Œåªä¿ç•™åè¯ã€åŠ¨è¯ã€å½¢å®¹è¯ï¼Œå¹¶å°†ç»“æœå­˜å‚¨åˆ°æ–°çš„ 'text1_cut' åˆ—ä¸­
data['text1_cut'] = data['text1'].apply(lambda x: jieba_cut_filtered(str(x)))

# æŸ¥çœ‹åˆ†è¯åçš„æ•°æ®
data[['text1', 'text1_cut']]
```

## step7 ç”¨åœç”¨è¯è¡¨å»é™¤åœç”¨è¯

```python
# è¯»å–åœç”¨è¯æ–‡ä»¶å¹¶å°†å…¶åˆå¹¶åˆ°ä¸€ä¸ªåœç”¨è¯é›†åˆä¸­
def load_stopwords(files):
    stopwords = set()
    for file in files:
        with open(file, 'r', encoding='utf-8') as f:
            words = f.read().splitlines()
            stopwords.update(words)
    return stopwords

# åœç”¨è¯æ–‡ä»¶è·¯å¾„
stopword_files = [
    '/kaggle/input/stopwords-list/cn_stopwords.txt',
    '/kaggle/input/stopwords-list/hit_stopwords.txt',
    '/kaggle/input/stopwords-list/baidu_stopwords.txt',
    '/kaggle/input/stopwords-list/scu_stopwords.txt'
]

# åŠ è½½åœç”¨è¯
stopwords = load_stopwords(stopword_files)

# å®šä¹‰å‡½æ•°æ¥ç§»é™¤åœç”¨è¯
def remove_stopwords(text, stopwords):
    words = text.split()  # å°†åˆ‡å¥½çš„è¯åˆ†å¼€
    cleaned_words = [word for word in words if word not in stopwords]  # è¿‡æ»¤åœç”¨è¯
    return ' '.join(cleaned_words)

# å¯¹ 'text1_cut' åˆ—è¿›è¡Œåœç”¨è¯æ¸…æ´—ï¼Œå¹¶å­˜å‚¨åˆ°æ–°çš„ 'text1_cleaned' åˆ—ä¸­
data['text1_cleaned'] = data['text1_cut'].apply(lambda x: remove_stopwords(str(x), stopwords))

# æŸ¥çœ‹æ¸…æ´—åçš„æ•°æ®
data[['text1_cut', 'text1_cleaned']].head()
```

## step8 æ„å»ºè‡ªå·±çš„è¯å…¸æŸ¥çœ‹è¯é¢‘

```python
import pandas as pd
from collections import Counter

# å°†æ‰€æœ‰æ–‡æœ¬åˆå¹¶å¹¶åˆ†å‰²æˆè¯æ±‡åˆ—è¡¨
all_words = ' '.join(data['text1_cleaned']).split()

# ç»Ÿè®¡è¯é¢‘
word_freq = Counter(all_words)

# å°†è¯é¢‘ç»“æœè½¬æ¢ä¸º DataFrame
word_freq_df = pd.DataFrame(word_freq.items(), columns=['word', 'frequency'])

# å°†ç»“æœä¿å­˜ä¸º Excel æ–‡ä»¶ï¼ˆç§»é™¤ encoding å‚æ•°ï¼‰
output_file = '/kaggle/working/word_frequency.xlsx'
word_freq_df.to_excel(output_file, index=False)

print(f"è¯é¢‘ç»Ÿè®¡å·²ä¿å­˜åˆ°: {output_file}")
```

## step10 å¹²å‡€æ•°æ®è¡¨

```python
import pandas as pd

# è¯»å–è¯é¢‘æ–‡ä»¶
word_freq_df = pd.read_excel('/kaggle/working/word_frequency.xlsx')

# è·å–è¯é¢‘å°äº 5çš„è¯
low_freq_words = word_freq_df[word_freq_df['frequency'] < 5]['word'].tolist()

# äººä¸ºè§„å®šçš„åœç”¨è¯
manual_stopwords = ['è±†ç“£', 'è¯´', 'åš', 'æƒ³', 'æ¥¼ä¸»', 'å§å¦¹']

# åˆå¹¶ä½é¢‘è¯å’Œäººä¸ºè§„å®šçš„åœç”¨è¯ï¼Œç»„æˆæ–°çš„åœç”¨è¯åˆ—è¡¨
new_stopwords = set(low_freq_words + manual_stopwords)

# å®šä¹‰æ¸…æ´—å‡½æ•°
def clean_text(text):
    words = text.split()  # ä»¥ç©ºæ ¼åˆ†éš”è¯
    cleaned_words = [word for word in words if word not in new_stopwords]  # å»é™¤åœç”¨è¯
    cleaned_text = ' '.join(cleaned_words)

    # å¤„ç†åŒä¹‰è¯
    cleaned_text = cleaned_text.replace('çˆ¸çˆ¸å¦ˆå¦ˆ', 'çˆ¶æ¯').replace('çˆ¸å¦ˆ', 'çˆ¶æ¯').replace('çˆ¹å¦ˆ', 'çˆ¶æ¯')
    cleaned_text = cleaned_text.replace('å¦ˆå¦ˆ', 'å¦ˆ').replace('æ¯äº²', 'å¦ˆ').replace('æˆ‘å¦ˆå¦ˆ', 'å¦ˆ').replace('æˆ‘å¦ˆ', 'å¦ˆ').replace('æˆ‘æ¯äº²', 'å¦ˆ')
    cleaned_text = cleaned_text.replace('è€äºº', 'è€å¹´äºº')
    cleaned_text = cleaned_text.replace('å­©å­', 'å­å¥³')
    cleaned_text = cleaned_text.replace('çˆ¸çˆ¸', 'çˆ¸').replace('æˆ‘çˆ¹', 'çˆ¸').replace('æˆ‘çˆ¸çˆ¸', 'çˆ¸').replace('æˆ‘çˆ¶äº²', 'çˆ¸').replace('çˆ¹', 'çˆ¸').replace('çˆ¶äº²', 'çˆ¸')
    
    return cleaned_text

# åº”ç”¨æ¸…æ´—å‡½æ•°ï¼Œç”Ÿæˆæ–°çš„ 'text2_cleaned' åˆ—
data['text2_cleaned'] = data['text1_cleaned'].apply(clean_text)

# æŸ¥çœ‹æ¸…æ´—åçš„æ•°æ®
print(data[['text1_cleaned', 'text2_cleaned']].head())
```
