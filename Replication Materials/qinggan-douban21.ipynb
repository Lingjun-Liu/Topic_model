## 2 读取词典，选词典变量

```python
import pandas as pd

df = pd.read_excel('/kaggle/input/qinggan-7/.xlsx')
df.head(10)
```

```python
#我们暂时只使用['词语', '词性种类', '词义数', '词义序号', '情感分类', '强度', '极性']
df = df[['词语', '词性种类', '词义数', '词义序号', '情感分类', '强度', '极性']]
df.head()
```

## 3 整理情绪词语列表

```python
#按照七大情绪划分
kuaile = []
anxin = []
zunjing = []
zanyang = []
xiangxin = []
xiai = []
zhuyuan = []
fennu = []
beishang = []
shiwang = []
jiu = []
si = []
huang = []
kongju = []
xiu = []
fanmen = []
zengwu = []
bianze = []
jidu = []
huaiyi = []
jingqi = []



for idx, row in df.iterrows():
    if row['情感分类'] in ['PA']:
        kuaile.append(row['词语'])
    if row['情感分类'] in ['PE']:
        anxin.append(row['词语'])   
    if row['情感分类'] in ['PD']:
        zunjing.append(row['词语']) 
    if row['情感分类'] in ['PH']:
        zanyang.append(row['词语']) 
    if row['情感分类'] in ['PG']:
        xiangxin.append(row['词语']) 
    if row['情感分类'] in ['PB']:
        xiai.append(row['词语'])
    if row['情感分类'] in ['PK']:
        zhuyuan.append(row['词语']) 
    if row['情感分类'] in ['NA']:
        fennu.append(row['词语'])         
    if row['情感分类'] in ['NB']:
        beishang.append(row['词语'])
    if row['情感分类'] in ['NJ']:
        shiwang.append(row['词语'])       
    if row['情感分类'] in ['NH']:
        jiu.append(row['词语'])
    if row['情感分类'] in ['PF']:
        si.append(row['词语'])     
    if row['情感分类'] in ['NI']:
        huang.append(row['词语'])
    if row['情感分类'] in ['NC']:
        kongju.append(row['词语'])
    if row['情感分类'] in ['NG']:
        xiu.append(row['词语'])          
    if row['情感分类'] in ['NE']:
        fanmen.append(row['词语'])
    if row['情感分类'] in ['ND']:
        zengwu.append(row['词语'])
    if row['情感分类'] in ['NN']:
        bianze.append(row['词语'])
    if row['情感分类'] in ['NK']:
        jidu.append(row['词语'])
    if row['情感分类'] in ['NL']:
        huaiyi.append(row['词语'])   
    if row['情感分类'] in ['PC']:
        jingqi.append(row['词语'])       
        
Positive = kuaile +anxin+zunjing+zanyang+xiangxin+xiai+zhuyuan+si+jingqi
Negative = fennu+beishang+shiwang+jiu+huang+kongju+xiu+fanmen+zengwu+bianze+jidu+huaiyi

print('情绪词语列表整理完成')     
```

## 4 情绪计算函数

```python
import jieba  # 导入 jieba 库
import pandas as pd

# 假设您已经在函数外定义了以下情感词汇列表
# kuaile, anxin, zunjing, ..., jingqi
# 以及 Positive 和 Negative 列表

def emotion_caculate(text):
    # 初始化情感和情绪计数变量，避免与情感列表名称冲突
    positive_count = 0
    negative_count = 0
    kuaile_count = 0
    anxin_count = 0
    zunjing_count = 0
    zanyang_count = 0
    xiangxin_count = 0
    xiai_count = 0
    zhuyuan_count = 0
    fennu_count = 0
    beishang_count = 0
    shiwang_count = 0
    jiu_count = 0
    si_count = 0
    huang_count = 0
    kongju_count = 0
    xiu_count = 0
    fanmen_count = 0
    zengwu_count = 0
    bianze_count = 0
    jidu_count = 0
    huaiyi_count = 0
    jingqi_count = 0

    # 将输入文本切分为词语列表
    wordlist = jieba.lcut(text)
    wordset = set(wordlist)  # 去重

    for word in wordset:
        freq = wordlist.count(word)
        if word in Positive:
            positive_count += freq
        if word in Negative:
            negative_count += freq
        if word in kuaile:
            kuaile_count += freq
        if word in anxin:
            anxin_count += freq
        if word in zunjing:
            zunjing_count += freq  
        if word in zanyang:
            zanyang_count += freq
        if word in xiangxin:
            xiangxin_count += freq
        if word in xiai:
            xiai_count += freq
        if word in zhuyuan:
            zhuyuan_count += freq
        if word in fennu:
            fennu_count += freq
        if word in beishang:
            beishang_count += freq
        if word in shiwang:
            shiwang_count += freq
        if word in jiu:
            jiu_count += freq
        if word in si:
            si_count += freq
        if word in huang:
            huang_count += freq
        if word in kongju:
            kongju_count += freq
        if word in xiu:
            xiu_count += freq
        if word in fanmen:
            fanmen_count += freq
        if word in zengwu:
            zengwu_count += freq
        if word in bianze:
            bianze_count += freq
        if word in jidu:
            jidu_count += freq
        if word in huaiyi:
            huaiyi_count += freq
        if word in jingqi:
            jingqi_count += freq

    # 汇总情感情绪分析结果并以字典形式存储
    emotion_info = {
        'length': len(wordlist),
        'positive': positive_count,
        'negative': negative_count,
        'kuaile': kuaile_count,
        'anxin': anxin_count,
        'zunjing': zunjing_count,
        'zanyang': zanyang_count,
        'xiangxin': xiangxin_count,
        'xiai': xiai_count,
        'zhuyuan': zhuyuan_count,
        'fennu': fennu_count,
        'beishang': beishang_count,
        'shiwang': shiwang_count,
        'jiu': jiu_count,
        'si': si_count,
        'huang': huang_count,
        'kongju': kongju_count,
        'xiu': xiu_count,
        'fanmen': fanmen_count,
        'zengwu': zengwu_count,
        'bianze': bianze_count,
        'jidu': jidu_count,
        'huaiyi': huaiyi_count,
        'jingqi': jingqi_count
    }

    # 返回包含情感情绪分析结果的 Pandas Series
    indexs = ['length', 'positive', 'negative','kuaile', 'anxin', 'zunjing', 'zanyang', 'xiangxin', 'xiai', 'zhuyuan', 'fennu', 'beishang', 'shiwang', 'jiu', 'si', 'huang', 'kongju', 'xiu', 'fanmen', 'zengwu', 'bianze', 'jidu', 'huaiyi', 'jingqi']
    return pd.Series(emotion_info, index=indexs)

# 示例调用情感分析函数
emotion_caculate(text='这个国家再对这些制造假冒伪劣食品药品的人手软的话，那后果真的会相当糟糕。坐牢？从快判个死刑')
```

## 5 并行加速

```python
!pip3 install pandarallel
```

```python
from pandarallel import pandarallel
import time

# 并行初始化
pandarallel.initialize()

# 确保 'text' 列为字符串类型，并处理缺失值
weibo_df['text'] = weibo_df['text'].fillna('').astype(str)

# 开始计时
start = time.time()

# 并行执行情感分析
emotion_df = weibo_df['text'].parallel_apply(emotion_caculate)

# 结束计时
end = time.time()
print(end - start)

# 显示结果
emotion_df.head()
```

## 6 输出分析结果

```python
output_df = pd.concat([weibo_df, emotion_df], axis=1)
output_df.to_csv('/kaggle/working/stm_13_qinggan21.csv', index=False)
output_df.head()
```
